```python
#http://scikit-learn.org/stable/tutorial/statistical_inference/unsupervised_learning.html

from sklearn import cluster, datasets
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = datasets.load_iris()
X_iris = iris.data
y_iris = iris.target
```

```python
k_means = cluster.KMeans(n_clusters=3)
k_means.fit(X_iris)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
color_code = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k', 7:'w'}
color_labels = []
for i in k_means.labels_[:]:
    color_labels.append(color_code[i])

xs = X_iris[:,0]
ys = X_iris[:,1]
zs = X_iris[:,2]
ls = X_iris[:,3]
cs = color_labels
ax.scatter(xs, ys, zs, s=ls, c=cs)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
